```python
def create_features(df,lags=17):
    df=df.copy()
    for lag in range(1,lags+1):
        df[f'lag_{lag}']=df.groupby('Work center')['ILP Tonnage Weight'].shift(lag)
    df['dayofweek']=df['Posting Date'].dt.dayofweek
    return df.dropna()
df_features=create_features(df_daily)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split

models={}
predictions=[]

work_centers=df_features["Work center"].unique()
future_dates=pd.date_range(start=df_features['Posting Date'].max()+pd.Timedelta(days=1),periods=17)

for wc in work_centers:
    wc_df=df_features[df_features['Work center']==wc].copy()
    X=wc_df.drop(columns=['Posting Date','ILP Tonnage Weight','Work center'])
    y=wc_df['ILP Tonnage Weight']

    model=XGBRegressor(n_estimators=3000, learning_rate=0.1, max_depth=3, random_state=42)
    model.fit(X,y)
    models[wc]=model

    last_row=wc_df.sort_values('Posting Date').iloc[-1:]
    preds=[]

    for i in range(17):
        features=last_row.copy()
        for lag in range(17,1,-1):
            features[f'lag_{lag}']=features[f'lag_{lag-1}']
            features['lag_1']=last_row['ILP Tonnage Weight'].values[0] if i==0 else preds[-1]
            #features['dayofweek'] = (last_row['Posting Date'].dt.dayofweek.values[0] + i + 1) % 7
            pred=model.predict(features.drop(columns=['ILP Tonnage Weight','Posting Date','Work center']))
            preds.append(pred[0])
    for d,p in zip(future_dates,preds):
        predictions.append({'Posting Date': d, 'Work center': wc, 'Predicted ILP Tonnage Weight': p})

df_preds = pd.DataFrame(predictions)

# Overall daily prediction
df_overall = df_preds.groupby('Posting Date')['Predicted ILP Tonnage Weight'].sum().reset_index()

df_sum=df_overall['Predicted ILP Tonnage Weight'].sum()
```
