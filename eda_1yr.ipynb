```python
import pandas as pd
import os

# Folder paths
input_folder = r"C:\Users\chenn\Desktop\New folder\prod'24"
output_folder = r"C:\Users\chenn\Desktop\New folder\2024_prod_data"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Columns you want to keep
columns_to_keep = ['Posting Date', 'Work center', 'Customer Segmant','Thickness (mm)','Grade Series number', 'ILP Usage', 'ILP Weight','Output Weight (101)']

# Loop through Excel files
for filename in os.listdir(input_folder):
    if filename.endswith(".XLSX"):
        print(f"Reading Excel file: {filename}")
        excel_path = os.path.join(input_folder, filename)
        excel_file = pd.ExcelFile(excel_path)

        for sheet_name in excel_file.sheet_names:
            try:
                df = pd.read_excel(excel_path, sheet_name=sheet_name)

                # Keep only selected columns (ignore missing columns gracefully)
                df_filtered = df.loc[:, df.columns.isin(columns_to_keep)]

                # Construct CSV file name
                base_name = os.path.splitext(filename)[0]
                csv_filename = f"{base_name}_{sheet_name}.csv"
                csv_path = os.path.join(output_folder, csv_filename)

                # Save filtered DataFrame
                df_filtered.to_csv(csv_path, index=False, encoding='utf-8')
                print(f"Saved: {csv_filename}")
            except Exception as e:
                print(f"Error in {filename} - Sheet: {sheet_name} -> {e}")
```

```python
#xgboost second model train test
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score

# STEP 1: Preprocess and Aggregate Data Per Day
df = df_filtered.copy()
df['Posting Date'] = pd.to_datetime(df['Posting Date'])

# Label encode categorical features
for col in ['Customer Segmant', 'Grade Series number']:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# Group per day
df_daily = df.groupby('Posting Date').agg({
    'ILP Tonnage Weight': 'sum',
    'Thickness (mm)': 'mean',
    'Output Weight (101)': 'sum',
    'Customer Segmant': 'mean',
    'Grade Series number': 'mean',
    'Work center': 'nunique'
}).reset_index()

# STEP 2: Create lag features
for lag in range(1, 15):
    df_daily[f'lag_{lag}'] = df_daily['ILP Tonnage Weight'].shift(lag)

df_daily = df_daily.dropna().reset_index(drop=True)

# STEP 3: Train/Test Split (last 20% for test)
X = df_daily.drop(columns=['Posting Date', 'ILP Tonnage Weight'])
y = df_daily['ILP Tonnage Weight']

split_point = int(len(df_daily) * 0.8)   # 80/20 split
X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

# STEP 4: Train Model
model = XGBRegressor(n_estimators=1000, learning_rate=0.19, max_depth=4, random_state=42)
model.fit(X_train, y_train)

# STEP 5: Evaluate
y_pred = model.predict(X_test)

rmse = mean_squared_error(y_test, y_pred) ** 0.5
r2 = r2_score(y_test, y_pred)

print("Test RMSE:", rmse)
print("Test R²:", r2)

# STEP 6: Forecast Future 30 Days (using full data for retraining)
model.fit(X, y)  # retrain on all data

future_preds = []
last_row = df_daily.iloc[-1:].copy()

for _ in range(30):
    new_input = last_row.drop(columns=['Posting Date', 'ILP Tonnage Weight']).copy()

    pred = model.predict(new_input)[0]
    future_preds.append(pred)

    # Slide window for next prediction
    new_row = last_row.copy()
    for lag in range(14, 1, -1):
        new_row[f'lag_{lag}'] = new_row[f'lag_{lag - 1}']
    new_row['lag_1'] = pred
    new_row['ILP Tonnage Weight'] = pred
    new_row['Posting Date'] = new_row['Posting Date'] + pd.Timedelta(days=1)

    last_row = new_row

# STEP 7: Output Future Forecast
future_dates = pd.date_range(start=df_daily['Posting Date'].max() + pd.Timedelta(days=1), periods=30)
df_forecast = pd.DataFrame({
    'Posting Date': future_dates,
    'Forecasted ILP Tonnage Weight': future_preds
})

print("30-Day Forecast Total:", df_forecast['Forecasted ILP Tonnage Weight'].sum())
print(df_forecast)
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# --------------------------
# STEP 1: Prepare data
# --------------------------
# Assuming df_daily has columns: ['Posting Date', 'ILP Tonnage Weight', lag_1 ... lag_n]

X = df_daily.drop(columns=['Posting Date', 'ILP Tonnage Weight'])
y = df_daily['ILP Tonnage Weight']

# Train-Test Split (time-based split is better for time series, but here simple split)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# --------------------------
# STEP 2: Train CatBoost
# --------------------------
model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=5,
    verbose=0,
    random_state=42
)

model.fit(X_train, y_train)

# --------------------------
# STEP 3: Evaluate on Test Set
# --------------------------
y_pred = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Test RMSE: {rmse:.4f}")
print(f"Test R²: {r2:.4f}")

# --------------------------
# STEP 4: Future Prediction (Next 30 Days)
# --------------------------
future_preds = []
last_row = df_daily.iloc[-1:].copy()

for _ in range(30):
    input_data = last_row.drop(columns=['Posting Date', 'ILP Tonnage Weight']).copy()
    pred = model.predict(input_data)[0]
    future_preds.append(pred)

    # Slide lag values
    for lag in range(14, 1, -1):
        last_row[f'lag_{lag}'] = last_row[f'lag_{lag - 1}']
    last_row['lag_1'] = pred
    last_row['ILP Tonnage Weight'] = pred
    last_row['Posting Date'] += pd.Timedelta(days=1)

# --------------------------
# STEP 5: Output Forecast
# --------------------------
future_dates = pd.date_range(
    start=df_daily['Posting Date'].max() + pd.Timedelta(days=1),
    periods=30
)

df_forecast = pd.DataFrame({
    'Posting Date': future_dates,
    'Forecasted ILP Tonnage Weight': future_preds
})

print(df_forecast)
print("Total Forecasted ILP Weight:", df_forecast['Forecasted ILP Tonnage Weight'].sum())
```
